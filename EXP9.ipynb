```python
# ==========================================
# System Login Log Brute-Force Detection
# Google Colab Compatible
# ==========================================

print("=== Brute-Force Intrusion Detection ===\n")

# Simulated login log entries
logs = [
    "2026-07-23 09:00:01 192.168.1.10 SUCCESS",
    "2026-07-23 09:01:10 192.168.1.20 FAILED",
    "2026-07-23 09:01:20 192.168.1.20 FAILED",
    "2026-07-23 09:01:30 192.168.1.20 FAILED",
    "2026-07-23 09:01:40 192.168.1.20 FAILED",
    "2026-07-23 09:01:50 192.168.1.20 FAILED",
    "2026-07-23 09:02:00 192.168.1.20 SUCCESS",
    "2026-07-23 09:03:10 192.168.1.30 FAILED",
    "2026-07-23 09:03:20 192.168.1.30 FAILED",
    "2026-07-23 09:04:10 192.168.1.40 SUCCESS"
]

# Threshold for brute-force detection
THRESHOLD = 5

failed_attempts = {}

# Analyze logs
for entry in logs:
    parts = entry.split()

    ip = parts[2]
    status = parts[3]

    if status == "FAILED":
        failed_attempts[ip] = failed_attempts.get(ip, 0) + 1
    elif status == "SUCCESS":
        failed_attempts[ip] = 0

# Display results
print("Failed Login Attempts")
print("-" * 35)

for ip, count in failed_attempts.items():
    print(f"{ip} : {count} failed attempt(s)")

print("\nPossible Brute-Force Attacks")
print("-" * 35)

attack_found = False

for ip, entry in [(line.split()[2], line.split()[3]) for line in logs]:
    pass

# Recalculate maximum consecutive failures
current = {}

for line in logs:
    parts = line.split()
    ip = parts[2]
    status = parts[3]

    if status == "FAILED":
        current[ip] = current.get(ip, 0) + 1
        if current[ip] >= THRESHOLD:
            print(f"ALERT: {ip} made {current[ip]} consecutive failed login attempts.")
            attack_found = True
    else:
        current[ip] = 0

if not attack_found:
    print("No brute-force attack detected.")

print("\nAnalysis Completed.")
```
