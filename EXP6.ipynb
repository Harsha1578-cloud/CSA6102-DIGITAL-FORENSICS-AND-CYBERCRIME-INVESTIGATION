```python
# ==========================================
# Bit-Stream (Forensic) Copy with Hash Verification
# Google Colab Compatible
# ==========================================

import hashlib
import shutil

print("=== Bit-Stream Copy and Integrity Verification ===\n")

# Create a sample evidence file
original_file = "evidence.txt"

with open(original_file, "w") as f:
    f.write("This is sample digital evidence used for forensic analysis.")

# Name of copied file
copy_file = "evidence_copy.txt"

# Create a bit-stream copy
shutil.copyfile(original_file, copy_file)

# Function to calculate SHA-256 hash
def calculate_hash(filename):
    sha256 = hashlib.sha256()

    with open(filename, "rb") as file:
        while True:
            data = file.read(4096)
            if not data:
                break
            sha256.update(data)

    return sha256.hexdigest()

# Calculate hashes
original_hash = calculate_hash(original_file)
copy_hash = calculate_hash(copy_file)

# Display results
print("Original File :", original_file)
print("Copied File   :", copy_file)

print("\nSHA-256 Hash of Original File:")
print(original_hash)

print("\nSHA-256 Hash of Copied File:")
print(copy_hash)

# Verify integrity
print("\nIntegrity Verification")
print("-" * 30)

if original_hash == copy_hash:
    print("SUCCESS: The forensic copy is identical to the original.")
    print("Integrity Verified.")
else:
    print("ERROR: Hash values do not match.")
    print("Integrity Verification Failed.")
```
