```python
# ============================================
# Deleted File Recovery Simulation
# Google Colab Compatible
# ============================================

print("Deleted File Recovery Simulation\n")

# Simulated storage (byte stream)
storage = (
    b"\x00\x01\x02\x03"
    + b"\xFF\xD8\xFF\xE0" + b"JPEG_IMAGE_DATA"      # JPEG Signature
    + b"\x11\x22\x33\x44\x55"
    + b"\x89PNG\r\n\x1A\n" + b"PNG_IMAGE_DATA"      # PNG Signature
    + b"\x66\x77\x88"
    + b"%PDF-" + b"PDF_DOCUMENT_DATA"              # PDF Signature
    + b"\x99\xAA\xBB"
    + b"PK\x03\x04" + b"ZIP_ARCHIVE_DATA"          # ZIP Signature
    + b"\xCC\xDD"
    + b"GIF89a" + b"GIF_IMAGE_DATA"                # GIF Signature
)

# Known file signatures
signatures = {
    "JPEG": b"\xFF\xD8\xFF",
    "PNG": b"\x89PNG\r\n\x1A\n",
    "PDF": b"%PDF-",
    "ZIP": b"PK\x03\x04",
    "GIF": b"GIF89a"
}

print("Scanning simulated storage...\n")

found = False

for file_type, signature in signatures.items():
    start = 0
    while True:
        pos = storage.find(signature, start)
        if pos == -1:
            break

        found = True
        print(f"Possible {file_type} file found")
        print(f"Offset : {pos}")
        print(f"Signature : {signature}")
        print("-" * 40)

        start = pos + 1

if not found:
    print("No recognizable deleted file signatures found.")

print("\nRecovery scan completed.")
```
