```python
import os
import time
from datetime import datetime

# File name
filename = "evidence.txt"

# Step 1: Create the file
with open(filename, "w") as file:
    file.write("Original evidence data.\n")

print("File created successfully.")

# Function to display file timestamps
def display_metadata(file):
    stats = os.stat(file)

    print("\n------ File Metadata ------")
    print("Created Time :",
          datetime.fromtimestamp(stats.st_ctime))
    print("Modified Time:",
          datetime.fromtimestamp(stats.st_mtime))
    print("Accessed Time:",
          datetime.fromtimestamp(stats.st_atime))
    print("----------------------------")

# Display original timestamps
print("\nInitial Metadata:")
display_metadata(filename)

# Wait for a few seconds
print("\nWaiting for 5 seconds...")
time.sleep(5)

# Step 2: Modify the file
with open(filename, "a") as file:
    file.write("This line represents a digital trace.\n")

print("\nFile modified.")

# Display updated timestamps
print("\nUpdated Metadata:")
display_metadata(filename)

print("\nLocard's Exchange Principle Demonstrated!")
print("The modification changed the file metadata, leaving a digital trace.")
```
