```python
import hashlib

# Step 1: Create an evidence file automatically
filename = "evidence.txt"

with open(filename, "w") as file:
    file.write("Digital Evidence\n")
    file.write("Case ID: DF001\n")
    file.write("Collected Date: 16-07-2026\n")

# Function to calculate SHA-256 hash
def calculate_hash(file_name):
    sha256 = hashlib.sha256()

    with open(file_name, "rb") as file:
        while True:
            data = file.read(4096)
            if not data:
                break
            sha256.update(data)

    return sha256.hexdigest()

# Step 2: Generate the original hash
original_hash = calculate_hash(filename)

print("Evidence File Created Successfully.")
print("\nOriginal SHA-256 Hash:")
print(original_hash)

# Step 3: Verify the file by calculating the hash again
current_hash = calculate_hash(filename)

print("\nCurrent SHA-256 Hash:")
print(current_hash)

# Step 4: Compare hashes
if original_hash == current_hash:
    print("\nEvidence Verified")
    print("Chain of Custody Maintained.")
else:
    print("\nEvidence Modified")
    print("Chain of Custody Broken.")
```
