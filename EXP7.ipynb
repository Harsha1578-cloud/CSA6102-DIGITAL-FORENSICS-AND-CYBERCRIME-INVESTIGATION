```python
# ==========================================
# File Type Identification using Magic Number
# Google Colab Compatible
# ==========================================

print("=== File Type Identification ===\n")

# Create sample files with magic numbers
samples = {
    "sample.jpg": b"\xFF\xD8\xFF\xE0",
    "sample.png": b"\x89PNG\r\n\x1A\n",
    "sample.pdf": b"%PDF-",
    "sample.zip": b"PK\x03\x04",
    "sample.gif": b"GIF89a"
}

# Write sample files
for filename, signature in samples.items():
    with open(filename, "wb") as f:
        f.write(signature)
        f.write(b"SampleData")

# Known magic numbers
magic_numbers = {
    b"\xFF\xD8\xFF": "JPEG Image",
    b"\x89PNG\r\n\x1A\n": "PNG Image",
    b"%PDF-": "PDF Document",
    b"PK\x03\x04": "ZIP Archive",
    b"GIF87a": "GIF Image",
    b"GIF89a": "GIF Image"
}

# Function to identify file type
def identify_file(filename):
    with open(filename, "rb") as f:
        header = f.read(8)

    for magic, filetype in magic_numbers.items():
        if header.startswith(magic):
            return filetype

    return "Unknown File Type"

# Analyze files
for filename in samples:
    print(f"{filename} --> {identify_file(filename)}")

print("\nAnalysis Completed.")
```
