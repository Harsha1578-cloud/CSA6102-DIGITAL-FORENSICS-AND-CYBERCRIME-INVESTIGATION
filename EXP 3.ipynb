```python
sender = input("Sender: ")
subject = input("Subject: ")
body = input("Body: ")

score = 0


if "@" not in sender:
    score += 1

if "urgent" in subject.lower() or "verify" in subject.lower():
    score += 1

if "click here" in body.lower() or "password" in body.lower():
    score += 1

if score >= 2:
    print("Phishing Email Detected")
else:
    print("Email Appears Safe")
```
