```python
# Digital Forensic Investigation Lifecycle Report Generator

print("=" * 60)
print("      DIGITAL FORENSIC INVESTIGATION REPORT")
print("=" * 60)

case_name = input("Enter Case Name: ")
investigator = input("Enter Investigator Name: ")
date = input("Enter Investigation Date: ")

print("\n" + "=" * 60)
print("CASE DETAILS")
print("=" * 60)
print("Case Name      :", case_name)
print("Investigator   :", investigator)
print("Date           :", date)

print("\n" + "=" * 60)
print("DIGITAL FORENSIC INVESTIGATION LIFECYCLE")
print("=" * 60)

stages = [
    ("1. Identification",
     "Identify the incident, devices, digital evidence, and scope of investigation."),

    ("2. Preservation",
     "Secure and preserve digital evidence without altering original data."),

    ("3. Collection",
     "Collect digital evidence using approved forensic methods and tools."),

    ("4. Examination & Analysis",
     "Examine, recover, analyze, and interpret digital evidence to identify facts."),

    ("5. Reporting & Presentation",
     "Prepare the forensic report and present findings with supporting evidence.")
]

for stage, description in stages:
    print("\n" + stage)
    print("-" * len(stage))
    print(description)

print("\n" + "=" * 60)
print("INVESTIGATION SUMMARY")
print("=" * 60)
print("Case Name    :", case_name)
print("Investigator :", investigator)
print("Status       : Investigation Completed")
print("Evidence     : Successfully Processed")
print("Report       : Generated Successfully")

print("\nThank you.")
```
