```python
# ==========================================
# Email Header Analysis - Spoofed Sender Detection
# Google Colab Compatible
# ==========================================

print("=== Email Header Analysis ===\n")

# Sample raw email header
email_header = """
From: support@paypal.com
Return-Path: attacker@gmail.com
Reply-To: scammer@yahoo.com
Received: from unknown (192.168.1.100)
Received: from mail.fake.com (203.0.113.10)
Subject: Verify Your Account
"""

# Extract header fields
fields = {}

for line in email_header.strip().split("\n"):
    if ":" in line:
        key, value = line.split(":", 1)
        fields.setdefault(key.strip(), []).append(value.strip())

# Display extracted fields
print("Extracted Header Fields")
print("-" * 35)

for key, values in fields.items():
    for value in values:
        print(f"{key}: {value}")

print("\nSpoofing Analysis")
print("-" * 35)

# Check Return-Path
if "From" in fields and "Return-Path" in fields:
    if fields["From"][0] != fields["Return-Path"][0]:
        print("Suspicious: From address differs from Return-Path.")

# Check Reply-To
if "Reply-To" in fields and "From" in fields:
    if fields["Reply-To"][0] != fields["From"][0]:
        print("Suspicious: Reply-To differs from From address.")

# Check Received headers
if "Received" in fields:
    print(f"Number of Received headers: {len(fields['Received'])}")

    if len(fields["Received"]) < 2:
        print("Warning: Very few mail servers handled this email.")

print("\nResult:")
if (
    fields.get("From", [""])[0] != fields.get("Return-Path", [""])[0]
    or fields.get("Reply-To", [""])[0] != fields.get("From", [""])[0]
):
    print("Possible Email Spoofing Detected!")
else:
    print("No obvious spoofing indicators found.")

print("\nAnalysis Completed.")
```
