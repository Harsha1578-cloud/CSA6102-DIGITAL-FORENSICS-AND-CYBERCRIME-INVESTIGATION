```python
# DNS Lookup using Python
# Basic OSINT Reconnaissance Technique

import socket

print("===================================")
print("      DNS Lookup Tool (OSINT)")
print("===================================")

# Get domain name from user
domain = input("Enter the domain name (e.g., google.com): ")

try:
    # Perform DNS lookup
    ip_address = socket.gethostbyname(domain)

    print("\nDNS Lookup Successful!")
    print("----------------------------")
    print("Domain Name :", domain)
    print("IP Address  :", ip_address)

except socket.gaierror:
    print("\nError: Unable to resolve the domain name.")
```
